```python
import numpy as np
import matplotlib.pylab as py

def QuadAlphaPos(a,b,c): #this equation calculates the root using + over - with the standard formula
    discrim = b*b-4*a*c
    if discrim < 0:
        x = 'The root is complex'
    else:
        x = (-1*b + np.sqrt(b*b-4*a*c))/(2*a)
    return x

def QuadAlphaNeg(a,b,c): #this equation calculates the root using - over + with the standard formula
    discrim = b*b-4*a*c
    if discrim < 0:
        x = 'The root is complex'
    else:
        x = (-1*b - np.sqrt(b*b-4*a*c))/(2*a)
    return x

def QuadBetaPos(a,b,c): #this equation calculates the root using + over - with the altered formula
    discrim = b*b-4*a*c
    if discrim < 0:
        x = 'The root is complex'
    else:
        x = (-1*b + np.sqrt(b*b-4*a*c))/(2*a)
    return x

def QuadBetaNeg(a,b,c): #this equation calculates the root using - over + with the standard formula
    discrim = b*b-4*a*c
    if discrim < 0:
        x = 'The root is complex'
    else:
        x = (-1*b - np.sqrt(b*b-4*a*c))/(2*a)
    return x

AlphaPosArray=[]
AlphaNegArray=[]
BetaPosArray=[]
BetaNegArray=[]

def QuadFourCalc(a,b,c): #this formula calculates all four roots, stores them in an array, and prints them
    AlphaPos = QuadAlphaPos(a,b,c)
    AlphaPosArray.append(AlphaPos)
    AlphaNeg = QuadAlphaNeg(a,b,c)
    AlphaNegArray.append(AlphaNeg)
    BetaPos = QuadBetaPos(a,b,c)
    BetaPosArray.append(BetaPos)
    BetaNeg = QuadBetaNeg(a,b,c)
    BetaNegArray.append(BetaNeg)
    
    AlphaPosVal = abs(a*AlphaPos*AlphaPos+b*AlphaPos+c)
    AlphaNegVal = abs(a*AlphaNeg*AlphaNeg+b*AlphaNeg+c)
    BetaPosVal = abs(a*BetaPos*BetaPos+b*BetaPos+c)
    BetaNegVal = abs(a*BetaNeg*BetaNeg+b*BetaNeg+c)
    
    if AlphaPosVal > BetaPosVal: #this section of the code compares which value gives a function value closer to 0 and saves it as the "better solution"
        BetterPos = BetaPos
    elif AlphaPosVal < BetaPosVal:
        BetterPos = AlphaPos
    else:
        BetterPos = AlphaPos
        
    if AlphaNegVal > BetaNegVal:
        BetterNeg = BetaNeg
    elif AlphaNegVal < BetaNegVal:
        BetterNeg = AlphaNeg
    else:
        BetterNeg = AlphaNeg
    
    print("When a is", a,", b is", b,", and c is", c,":")
    print("The roots using the first formula are", AlphaPos, "and", AlphaNeg )
    print("The roots using the second formula are", BetaPos, "and", BetaNeg)
    print("The more precise solutions are", BetterPos, "and", BetterNeg, '\n')
    return 0
```

```python
a=1
b=1
c=1/10
x=[]
for i in range(1,20):
    QuadFourCalc(a,b,c)
    x.append(c)
    c=c/10
    
py.title('Solutions for Quadratic Forumla using as C is divided by 10')
py.xlabel('c')
py.ylabel('analytical solution')
py.plot(x,AlphaPosArray,'ro')
py.plot(x,AlphaNegArray,'bo')
py.plot(x,BetaPosArray,'go')
py.plot(x,BetaNegArray,'yo')
py.xscale('log')
py.show()
```
