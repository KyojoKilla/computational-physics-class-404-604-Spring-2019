```python
#Kyle Prerost
#PHYS 404
#HW 1
#January 23rd, 2019

import matplotlib.pylab as py
import numpy as np
x_axis=np.arange(-np.pi,np.pi,0.05)
#print(x_axis)

a=np.pi/2
for i in range(6):
    a=a+2*np.sin((2*i+1)*x_axis)/(2*i+1)
    #print(2*i+1)
#print(a)

b=np.pi/2
for i in range(11):
    b=b+2*np.sin((2*i+1)*x_axis)/(2*i+1)
    #print(2*i+1)
#print(b)

c=np.pi/2
for i in range(16):
    c=c+2*np.sin((2*i+1)*x_axis)/(2*i+1)
    #print(2*i+1)
#print(c)

py.plot(x_axis,a,'r')
py.plot(x_axis,b,'b')
py.plot(x_axis,c,'g')

py.show()
```
