```python
import matplotlib.pylab as py
import numpy as np

def fib_tab(n):
    table=[0]*(n+1)
    table[1] = 1
    
    for i in range(2,n+1):
        table[i]=table[i-1]+table[i-2]
        
    return table[n]
#This is the Tabulation Method (bottom up) for calculating Fibonacci numbers

x=np.arange(2,101,1)
#print(x) #remove '#' to see data points

y=[]
for i in range(0,99):
    y.append(fib_tab(x[i])/fib_tab(x[i-1]))
#print(y_axis) #remove '#' to see data points

py.title('Fibonacci Ratios')
py.xlabel('n')
py.ylabel('Fn/Fn-1')
py.plot(x,y,'r')
py.show()
```

```python
import matplotlib.pylab as py
import numpy as np

def fib_iter(n):
    if n == 0: return 0
    elif n == 1: return 1
    else: return fib_iter(n-1)+fib_iter(n-2)
#This is the iteration (recursion) method for calculating Fibonacci numbers

def fib_tab(n):
    table=[0]*(n+1)
    table[1] = 1
    
    for i in range(2,n+1):
        table[i]=table[i-1]+table[i-2]
        
    return table[n]
#This is the Tabulation Method (bottom up) for calculating Fibonacci numbers

def arrpow(arr, n):
    yarr=arr
    if n<1:
        print('n needs to be larger than 1')
    if n==1:
        return arr
    yarr = arrpow(arr, n//2)
    yarr = [[yarr[0][0]*yarr[0][0]+yarr[0][1]*yarr[1][0],yarr[0][0]*yarr[0][1]+yarr[0][1]*yarr[1][1]],
            [yarr[1][0]*yarr[0][0]+yarr[1][1]*yarr[1][0],yarr[1][0]*yarr[0][1]+yarr[1][1]*yarr[1][1]]]
    if n%2:
        yarr=[[yarr[0][0]*arr[0][0]+yarr[0][1]*arr[1][0],yarr[0][0]*arr[0][1]+yarr[0][1]*arr[1][1]],
            [yarr[1][0]*arr[0][0]+yarr[1][1]*arr[1][0],yarr[1][0]*arr[0][1]+yarr[1][1]*arr[1][1]]]
    return yarr

def fib_mat(n):
    arr=[[1,1],[1,0]]
    f=arrpow(arr,n-1)[0][0]
    return f
#This is the matrix method for calculating Fibonacci numbers

import time
import sys

f = open('data_hw2.txt', 'w')

x_iter=np.array([1,2,4,8,16,32,34]) #could only use these values as any more would crash the program
for i in range(0,7):
    print(x_iter[i], file = f)
    
x_tab=np.array([1,10,100,1000,10000,10000,100000])
for i in range(0,7):
    print(x_tab[i], file = f)
    
x_mat=np.array([10,100,1000,10000,10000,100000])
for i in range(0,6):
    print(x_mat[i], file = f)
    
start_time = time.time()
y_iter=[]
y_iter_time=[]
for i in range(0,7): 
    y_iter.append(fib_iter(x_iter[i]))
    y_iter_time.append(time.time()-start_time)
    print(y_iter_time[i], file = f)

start_time = time.time()
y_tab=[]
y_tab_time=[]
for i in range(0,7):
    y_tab.append(fib_tab(x_tab[i]))
    y_tab_time.append(time.time()-start_time)
    print(y_tab_time[i], file = f)

start_time = time.time()
y_mat=[]
y_mat_time=[]
for i in range(0,6):
    y_mat.append(fib_mat(x_mat[i]))
    y_mat_time.append(time.time()-start_time)
    print(y_mat_time[i], file = f)

#x1 = np.loadtxt(f)
#x1 = f.readline(1)

f.close()
#The above file exporting and importing did not quite work. I was unable to find any examples of how
#to construct this the portion of the code and did the plotting with the raw data. I would like to go
#over this in class.

import math as m
xlog_iter=[]
for i in range(0,7):
    xlog_iter.append(m.log(x_iter[i]))
xlog_tab=[]
for i in range(0,7):
    xlog_tab.append(m.log(x_tab[i]))
xlog_mat=[]
for i in range(0,6):
    xlog_mat.append(m.log(x_mat[i]))
    
ylog_iter=[]
for i in range(0,7):
    ylog_iter.append(m.log(y_iter[i]))
ylog_tab=[]
for i in range(0,7):
    ylog_tab.append(m.log(y_tab[i]))
ylog_mat=[]
for i in range(0,6):
    ylog_mat.append(m.log(y_mat[i]))
#The above code was used to take the log of the n and data we used.

py.title('Iteration Method')
py.xlabel('log(n)')
py.ylabel('time taken (log(sec))')
py.plot(xlog_iter,ylog_iter,'ro')
py.axis([0,12,0,50000])
py.show()

py.title('Tabulation Method')
py.xlabel('log(n)')
py.ylabel('time taken (log(sec))')
py.plot(xlog_tab,ylog_tab,'bo')
py.axis([0,12,0,50000])
py.show()

py.title('Matrix Method')
py.xlabel('log(n)')
py.ylabel('time taken (log(sec))')
py.plot(xlog_mat,ylog_mat,'go')
py.axis([0,12,0,50000])
py.show()
```
